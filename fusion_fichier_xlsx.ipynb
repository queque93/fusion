```python
import numpy as np
import pandas as pd
import glob
all_data = pd.DataFrame()
for f in glob.glob("*.xlsx"):
    df = pd.read_excel(f)
    all_data = all_data.append(df,ignore_index=True)

# now save the data frame

writer = pd.ExcelWriter('site_fusionner.xlsx')
all_data.to_excel(writer,'sheet1',index=False)
writer.save()    
```
