```python
import pandas as pd 
import os
import glob
```

```python

#Récupère la liste des fichiers spécifiés par le chemin au format liste

csv_files = glob.glob('C:/Users/HP/Documents/afp/Fichier1/*.csv')

#Afficher la liste des fichiers à lire

for a in csv_files:
    print(a)

#Préparez une liste pour ajouter le contenu du fichier csv

data_list = []

#Scannez la liste des fichiers à lire

for file in csv_files:
    data_list.append(pd.read_csv(file,encoding='latin-1',sep=';'))

#Combinez toutes les listes dans le sens des lignes
#axis=0:Rejoindre dans le sens de la ligne

df = pd.concat(data_list, axis=0)

# Trouver les colonnes qui sont vides 

cols_vides = [col for col in df.columns if df[col].isnull().all()]

# supprimer les colonnes vides  du tableau 

#df.drop(cols_vides, axis=1, inplace=True)
```
